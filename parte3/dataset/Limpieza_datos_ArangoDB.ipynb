# Limpieza de datos

### Insertamos el .csv

```python
import pandas as pd
Datos_Aeropuertos = pd.read_csv("airports.csv")
print("El dataset ha sido leido")
```

### Eliminamos las columnas que no se van a utilizar

```python
#Eliminar colummnas innecesarias en el estudio estadístico a realizar
Datos_Aeropuertos = Datos_Aeropuertos.drop(["id", "ident","home_link", "wikipedia_link", "keywords", "scheduled_service", "gps_code", "iata_code", "local_code"], axis = 1)

#Eliminar los NaN 
Datos_Aeropuertos = Datos_Aeropuertos.dropna().reset_index(drop=True)

#Creación de una columna con la _key para ArangoDB 
Datos_Aeropuertos['_key'] = Datos_Aeropuertos['name'] 

#Posicionar _key en primer lugar
cols = Datos_Aeropuertos.columns.tolist()
cols = cols[-1:] + cols[:-1]
Datos_Aeropuertos = Datos_Aeropuertos[cols]

#Seleccionar _key válida
Datos_Aeropuertos["_key"] = Datos_Aeropuertos._key.str.replace(' ', '')
#Datos_Aeropuertos.rename(columns={"name":'_key'}, inplace=True)
```

### Seleccionamos solo los datos de aeropuertos Europeos

```python
def continent_select(datos):
    dropRows = []
    for i in range(0, len(datos)):
        if("EU" not in datos.continent[i]):
            #print(datos.tpep_pickup_datetime[i])
            dropRows.append(i)
            
    datos_datetime = datos.drop(dropRows).reset_index(drop=True)   
    return datos_datetime
```

```python
 Datos_Aeropuertos = continent_select(Datos_Aeropuertos)
```

```python
Datos_Aeropuertos
```

### Seleccionamos solo los datos de aeropuertos Europeos y de tamaño grande

```python
def type_select(datos):
    dropRows = []
    for i in range(0, len(datos)):
        if("large_airport" not in datos.type[i]):
            #print(datos.tpep_pickup_datetime[i])
            dropRows.append(i)
            
    datos_datetime = datos.drop(dropRows).reset_index(drop=True)   
    return datos_datetime
```

```python
Datos_Aeropuertos = type_select(Datos_Aeropuertos)
```

```python
Datos_Aeropuertos.to_csv("Datos_Aeropuertos_Limpios.csv")
```

### Seleccionamos solo los datos de aeropuertos Europeos, de tamaño grande y Españoles

```python
def ESP_select(datos):
    dropRows = []
    for i in range(0, len(datos)):
        if("ES" not in datos.iso_country[i]):
            #print(datos.tpep_pickup_datetime[i])
            dropRows.append(i)
            
    datos_datetime = datos.drop(dropRows).reset_index(drop=True)   
    return datos_datetime
```

```python
Datos_Aeropuertos = ESP_select(Datos_Aeropuertos)
```
